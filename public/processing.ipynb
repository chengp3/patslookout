```python
import json
import os
from PIL import Image
import re

pathto_oldjson = '/Users/admin/Desktop/patslookout/public/photos.json'
pathto_newjson = '/Users/admin/Downloads/lookoutpics/photos.json'
pathto_pics = '/Users/admin/Downloads/lookoutpics/'

def find_number_at_end(s):
    match = re.search(r'\d+$', s)
    return match.group() if match else None

def get_name_without_number(s):
    match = re.search(r'\D+', s)
    return match.group() if match else None

def renameToLowercase(directory):
    for filename in os.listdir(directory):
        lowercase_filename = filename.lower()
        if lowercase_filename != filename:
            os.rename(
                os.path.join(directory, filename),
                os.path.join(directory, lowercase_filename)
            )
            print(f"Renamed '{filename}' to '{lowercase_filename}'")

def removeDS(path):
    li = os.listdir(path)
    for item in li:
        if item.startswith('.'):
            os.remove(path + item)
            print(f'Removed file: {path + item}')

def compareDirectories(dir1, dir2):
    if type(dir1) == str:
        files1 = set(os.listdir(dir1))
    else:
        files1 = set(dir1)
    if type(dir2) == str:
        files2 = set(os.listdir(dir2))
    else:
        files2 = set(dir2)
    
    uniques1 = list(files1 - files2)
    uniques2 = list(files1 - files2)

    if len(uniques1) != 0 or len(uniques2) !=0:
        print('File mismatches:')
        print(dir1, uniques1)
        print(dir2, uniques2)
        return False
    else: 
        return True
    
def file_in_json(json_files, new_obj, filename):
    if (filename in json_files and json_files[filename].width == new_obj.width and json_files[filename].height == new_obj.height):
        return True
    else:
        return False

def find_name(json_files, new_obj, filename):
    """
    file = new file
    check if exists in json. if so, check if width and height are also the same. if so, continue
    else need to find a new name. get num from end of filename. this is counter. increment by 1.
    while loop:            
        search for new filename in json. 
        if exists: check width and h
    """
    base_name = get_name_without_number(filename)
    counter = find_number_at_end(filename)
    counter = int(counter) if counter else 1
    if counter == 1:
        new_name = base_name
    else:
        new_name = base_name + str(counter)

    while (new_name in json_files):
        if file_in_json(json_files, new_obj, new_name):
            return -1
        else:
            counter += 1
            new_name = base_name + str(counter)

    return new_name

def resize_and_save_image(file_path, output_folder, size=(400, 300)):
    with Image.open(file_path) as img:
        img = img.resize(size, Image.ANTIALIAS)
        base_name = os.path.basename(file_path)
        save_path = os.path.join(output_folder, base_name)
        img.save(save_path)
        print(f'Processed and saved: {save_path}')

def main():
    oldfullsize = pathto_pics + 'oldfullsize/'
    oldthumbs = pathto_pics + 'oldthumbs/'
    new43 = pathto_pics + 'new43/'
    newfullsize = pathto_pics + 'newfullsize/'
    newthumbs = pathto_pics + 'newthumbs/'
    old43 = pathto_pics + 'old43/'

    alldirs = [oldfullsize, oldthumbs, new43, newfullsize, old43, newthumbs]
    for dir in alldirs:
        removeDS(dir)
        renameToLowercase(dir)

    # validate folders. should have a complete new43 folder to start
    removeDS(new43)
    removeDS(newfullsize)
    if not compareDirectories(new43, newfullsize):
        return
    if not compareDirectories(old43, oldfullsize):
        return
    if not compareDirectories(old43, oldthumbs):
        return
    print('Directories validated')
        
    # ingest old json. for every file
    with open(pathto_oldjson, 'r') as file:
        data = json.load(file)
        json_files = json_files = {item['filename']: item for item in data}
        if not compareDirectories(oldfullsize, json_files):
            return
        print('json loaded')
        
        new_files = os.listdir(new43)

        for file in new_files:
            new_obj = {
                'filename': file,
                'width': 0,
                'height': 0,
                'title': "",
                "tags": []
            }
            with Image.open(new43 + file) as img:
                new_obj['width'], new_obj['height'] = img.width, img.height
        
            og_base_name, extension = os.path.splitext(file)
            base_name = find_name(json_files, new_obj, og_base_name)
            if base_name == -1: 
                continue
            else:
                new_name = base_name + extension

            if base_name != og_base_name:
                new_obj['filename'] = new_name
                os.rename(os.path.join(new43, file), os.path.join(new43, new_name))
                os.rename(os.path.join(newfullsize, file), os.path.join(newfullsize, new_name))
            
            json_files[new_name] = new_obj
            data.append(new_obj)
            resize_and_save_image(new43 + new_name, newthumbs, (400, 300))

        with open(pathto_newjson, 'w') as file:
            json.dump(data, file, indent=2)

main()
```
